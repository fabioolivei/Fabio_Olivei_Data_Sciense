# **Análise Exploratória de Dados (EDA)**

Aqui estão os principais gráficos gerados e insights obtidos:

**A IA generativa é uma prioridade em sua empresa?**

Utilizando um heatmap, observamos que a maioria das empresas ainda não considera a IA generativa uma prioridade.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações para a visualização
sns.set(style="white")  # Remove o grid padrão
plt.rcParams["figure.figsize"] = (20, 10)  # Aumentar largura para comportar melhor os rótulos

# Nome completo da coluna para AI Generativa
column_ai = "('P3_e ', 'AI Generativa é uma prioridade em sua empresa?')"

# Função para plotar gráfico de mapa de calor
def plot_individual_analysis_heatmap(data, column, title, xlabel='Porcentagem', ylabel='Respostas'):
    plt.figure(figsize=(16, 10))
    response_counts = data[column].value_counts()


    # Calcular porcentagens
    total = response_counts.sum()
    percentages = response_counts / total * 100

    # Verificar a soma das porcentagens
    sum_percentages = percentages.sum()
    print(f"Soma das porcentagens (verificação): {sum_percentages:.2f}%")

    # Criar o heatmap a partir dos dados de porcentagens
    heatmap_data = percentages.values.reshape(-1, 1)  # Redimensionar para 2D
    ax = sns.heatmap(heatmap_data, annot=True, fmt=".1f",  cmap="YlGnBu", cbar=False, yticklabels=response_counts.index)

    # Remover spines
    sns.despine(left=True, bottom=True)
    plt.title(f'Análise de Prioridade: {title}', fontsize=20, fontweight='bold')
    plt.xlabel(xlabel, fontsize=16)
    plt.ylabel(ylabel, fontsize=16)
    plt.xticks([], [])  # Remove os ticks no eixo X, pois não há categorias

    plt.show()

# Análise individual para a pergunta sobre AI Generativa com mapa de calor
plot_individual_analysis_heatmap(data, column_ai, 'AI Generativa é prioridade em sua empresa?', ylabel='Respostas')
```

* **Adoção Inicial:** A maioria das empresas (30.7%) está explorando a IA generativa, mas ela ainda não é uma prioridade máxima. Muitas empresas estão em fase de avaliação.
* **Baixa Prioridade:** Cerca de 28.8% das empresas afirmam que a IA generativa não é uma prioridade, sugerindo falta de interesse ou infraestrutura.
* **Planejamento Futuro:** 23.9% das empresas planejam priorizar a IA generativa nos próximos 2–4 anos, indicando um crescimento futuro.
* **Adoção Líder:** Apenas 12.3% colocam a IA generativa como a principal prioridade, investindo fortemente na tecnologia.
* **Desconhecimento:** 23.6% das empresas e colaboradores ainda não sabem opinar ou não têm visibilidade sobre soluções de IA generativa e LLMs, mostrando uma necessidade de maior conscientização e educação no setor.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Nome completo da coluna para tipos de uso de AI Generativa
column_uses = "('P3_f ', 'Tipos de uso de AI Generativa e LLMs na empresa')"

# Filtrar os dados para contar as respostas e pegar os top 5
response_counts = data[column_uses].value_counts().nlargest(5).dropna()  # Top 5 e remover valores nulos

# Calcular porcentagens
total_responses = response_counts.sum()
percentages = (response_counts / total_responses * 100).dropna()  # Remover valores nulos

# Verificar a soma das porcentagens
sum_percentages = percentages.sum()
print(f"Soma das porcentagens (verificação): {sum_percentages:.2f}%")

# Resumir manualmente os rótulos longos para o gráfico de linha (Top 5)
shortened_labels = [
    'Uso AI/LLMs',                    # Exemplo: Tipos de uso de AI Generativa e LLMs na empresa
    'Não sei opinar',                 # Exemplo: Não sei opinar sobre o uso de IA Generativa e LLMs
    'Colaboradores descentralizados', # Exemplo: Colaboradores usando AI generativa de forma independente
    'Melhorar produtos internos',     # Exemplo: AI Generativa e LLMs para melhorar produtos internos
    'Uso centralizado'                # Exemplo: Direcionamento centralizado do uso de AI generativa
]

# Função para plotar gráfico de linha com rótulos nas porcentagens e largura aumentada
def plot_simple_analysis_line_with_labels(response_counts, percentages, title, xlabel='Porcentagem', ylabel='Respostas'):
    plt.figure(figsize=(12,6))  # Aumentar a largura do gráfico para comportar os rótulos

    # Gráfico de linha
    plt.plot(percentages.values, marker='o', linestyle='-', color='green')

    # Adicionar rótulos nas porcentagens
    for i, percentage in enumerate(percentages.values):
        if pd.notna(percentage):  # Certifique-se de que o valor é válido
            plt.text(i, percentage, f'{percentage:.1f}%', ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

    # Configurar os ticks no eixo X com os índices dos rótulos
    plt.xticks(range(len(shortened_labels)), labels=shortened_labels, fontsize=12, rotation=25, ha='right')

    plt.title(f'Top 5 Usos: {title}', fontsize=20, fontweight='bold')
    plt.xlabel(xlabel, fontsize=16)
    plt.ylabel(ylabel, fontsize=16)
    plt.yticks(fontsize=14)

    # Remover os spines
    sns.despine(top=True, bottom=True)

    plt.tight_layout()  # Ajustar o layout para evitar sobreposição
    plt.show()

# Análise individual para a pergunta sobre Tipos de uso de AI Generativa com gráfico de linha e rótulos resumidos
plot_simple_analysis_line_with_labels(response_counts, percentages, 'Uso de AI Generativa e LLMs na empresa', ylabel='Respostas')
```

**Definições**:

* Uso AI/LLMs: Utilização de IA generativa e LLMs para diversas finalidades.

* Não sei opinar: Resposta neutra indicando desconhecimento sobre o tema.

* Colaboradores descentralizados: Colaboradores utilizando IA generativa de forma independente.

* Melhorar produtos internos: Aplicação de IA para otimizar produtos e ferramentas internas da empresa.

* Uso centralizado: Existe um direcionamento centralizado para que colaboradores utilizem soluções baseadas em AI Generativa (como o ChatGPT por exemplo), incluindo apoio nos custos das ferramentas.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Função para gerar gráficos de contagem com personalização
def gerar_graficos_com_personalizacao(dataframe):
    total = len(dataframe)

    for column in dataframe.columns:
        plt.figure(figsize=(10, 6))

        # Condição para aplicar cores personalizadas apenas em certas colunas
        if column in ['IA como principal frente do negócio', 'Uso e pago pelas soluções de AI',
                      'A empresa que trabalho paga pelas soluções de AI',
                      'Colaboradores usando AI de forma independente', 'AI Generativa e LLMs para melhorar produtos externos',
                      'AI Generativa e LLMs para melhorar produtos internos']:
            ax = sns.countplot(x=column, data=dataframe, palette=['black', 'darkgreen'])
        else:
            ax = sns.countplot(x=column, data=dataframe)

        # Definir cores personalizadas
        if column == ['IA como principal frente do negócio', 'Uso e pago pelas soluções de AI',
                      'A empresa que trabalho paga pelas soluções de AI',
                      'AI Generativa e LLMs para melhorar produtos externos',
                      'AI Generativa e LLMs para melhorar produtos internos']:
            # As cores específicas: verde escuro para 1 e preto para 0
            bar_colors = ['black' if x == 0 else 'darkgreen' for x in dataframe[column].unique()]
            for bar, color in zip(ax.patches, bar_colors):
                bar.set_facecolor(color)

            # Adicionar a legenda específica
            handles = [plt.Line2D([0], [0], color='black', lw=4),
                       plt.Line2D([0], [0], color='darkgreen', lw=4)]
            labels = ['0 = Não', '1 = Sim']
            ax.legend(handles, labels, title='Satisfeito')

        # Remover spines desnecessários
        ax.spines['top'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.spines['right'].set_visible(False)

        # Adicionar título
        ax.set_title(f'Distribuição de {column}')

        # Calcular e adicionar porcentagens sobre as barras
        for container in ax.containers:
            for bar in container:
                height = bar.get_height()
                percentage = f'{round(100 * height / total, 1)}%'
                ax.text(bar.get_x() + bar.get_width() / 2, height - (0.07 * height), percentage,
                        ha='center', va='top', color='white', fontweight='bold', fontsize=14)

        # Ajustar rótulos do eixo x
        plt.xticks(rotation=45, ha='right')

        # Exibir o gráfico
        plt.show()

# Chamada da função passando o dataframe completo, exceto a coluna "Tipo de uso de AI Generativa e LLMs"
gerar_graficos_com_personalizacao(IA.drop(columns=['Tipo de uso de AI Generativa e LLMs']))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular a matriz de correlação
# Excluindo a coluna com valores não numéricos para o cálculo da correlação
correlation_matrix = IA.drop(columns=['Tipo de uso de AI Generativa e LLMs']).corr()

# Criar o heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação das Variáveis de IA')
plt.show()
```

# **Conclusão**
A avaliação final da pesquisa do State of Data do Data Hackers sobre o uso de IA generativa e LLMs (Large Language Models) no Brasil mostra que, apesar do crescimento da adoção dessas tecnologias, elas ainda estão concentradas em grandes empresas e setores específicos. Essa constatação foi reforçada durante uma live deles realizada em parceria com a Thoughtworks, onde especialistas destacaram que a aplicação da IA generativa está sendo impulsionada principalmente por organizações que têm maior capacidade de investimento e infraestrutura tecnológica.

Na live, discutiu-se que essas grandes empresas, como bancos e fintechs, estão explorando a IA generativa para melhorar a eficiência operacional, personalizar o atendimento ao cliente e reduzir custos, principalmente em áreas como atendimento e detecção de fraudes. No entanto, para que esses projetos continuem a ser priorizados, é essencial que eles demonstrem um retorno claro sobre o investimento (ROI).

Isso alinha-se com os dados da pesquisa State of Data, que mostra que as empresas brasileiras estão mais cautelosas em adotar tecnologias de IA sem um caso de negócios bem fundamentado.

Outro ponto levantado durante a live foi a crescente pressão por resultados mensuráveis em termos financeiros, o que também reflete as descobertas da pesquisa. Empresas menores ou de setores menos maduros em termos de tecnologia ainda enfrentam desafios para justificar os custos elevados e a complexidade da implementação de IA generativa.

A expectativa é que, à medida que os custos de implementação diminuam e os benefícios sejam mais amplamente demonstrados, essa tecnologia se torne mais acessível a um número maior de empresas e setores no futuro. Porém como diriam muitos especialistas e líderes tecnológicos não há como  parar o avanço da IA, e o Brasil tem muito potencial para se posicionar com um player global em IA, especialmente em processamento de língua portuguesa e soluções para países em desenvolvimento.
