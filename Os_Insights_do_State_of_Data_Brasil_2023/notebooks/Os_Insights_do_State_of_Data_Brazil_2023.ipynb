# **Introdução**

O "**State of Data 2023**" é uma pesquisa anual conduzida pela Data Hackers, uma das maiores comunidades de profissionais de dados do Brasil. O relatório tem como objetivo mapear o cenário do mercado de dados no Brasil, fornecendo insights sobre as tendências, desafios, salários, tecnologias, ferramentas e o perfil dos profissionais que atuam nas áreas de ciência de dados, engenharia de dados, análise de dados, e outras funções relacionadas.


#### **Cenário**
Estamos embarcando em um projeto de ciência de dados com o objetivo de analisar e prever a satisfação dos profissionais de dados no Brasil em 2023. Esse projeto é de extrema importância para entender as dinâmicas do mercado de trabalho em dados, incluindo a identificação de fatores que influenciam a satisfação no trabalho, bem como fornecer insights valiosos para empresas e profissionais que atuam nesse setor.

#### **Objetivos do Projeto**
Nossos objetivos principais são os seguintes:

- Realizar uma análise exploratória dos dados (EDA) para entender as características e variáveis que compõem o mercado de dados no Brasil em 2023.

- Demonstrar as tendências, desafios e perspectivas de carreira para os profissionais de dados no Brasil, utilizando gráficos e análises correlacionais.

## Dicionário

1. **Idade**: Faixa etária do profissional de dados.
2. **Gênero**: Identificação de gênero do profissional (ex: Masculino, Feminino, Não binário, etc.).
3. **Vive no Brasil**: Indicação se o profissional reside no Brasil.
4. **UF**: Unidade Federativa (estado) onde o profissional reside.
5. **Ensino**: Nível de escolaridade do profissional (ex: Ensino Médio, Graduação, Pós-graduação, etc.).
6. **Formação**: Área de formação acadêmica do profissional (ex: Ciência de Dados, Engenharia, Estatística, etc.).
7. **Situação atual**: Situação atual do profissional no mercado de trabalho (ex: Empregado, Desempregado, Estudante, etc.).
8. **Cargo atual**: Cargo que o profissional ocupa atualmente (ex: Cientista de Dados, Analista de Dados, Engenheiro de Dados, etc.).
9. **Nível**: Nível de senioridade do profissional (ex: Júnior, Pleno, Sênior).
10. **Salário**: Faixa salarial do profissional.
11. **Satisfeito**: Indicador se o profissional está satisfeito com seu trabalho (variável alvo para o modelo preditivo).
12. **Mudaria de emprego**: Indicação se o profissional mudaria de emprego se tivesse a oportunidade.
13. **Tempo de experiência**: Tempo de experiência do profissional na área de dados.
14. **Atuação**: Área de atuação específica dentro do campo de dados (ex: Engenharia de Dados, Ciência de Dados, Análise de Dados).
15. **Forma de trabalho preferida**: Preferência do profissional em relação à forma de trabalho (ex: Remoto, Presencial, Híbrido).
16. **Forma de trabalho atual**: Forma de trabalho em que o profissional está atualmente inserido (ex: Remoto, Presencial, Híbrido).
17. **Qualidade dos gestores e líderes**: Percepção do profissional sobre a qualidade de seus gestores e líderes.
18. **Ambiente e clima de trabalho**: Percepção do profissional sobre o ambiente e o clima organizacional em sua empresa.
19. **Plano de carreira**: Existência e clareza de um plano de carreira na empresa do profissional.
20. **Maturidade da empresa**: Percepção sobre o nível de maturidade da empresa em que o profissional trabalha.
21. **Benefícios**: Benefícios oferecidos pela empresa ao profissional (ex: plano de saúde, vale alimentação, etc.).
22. **Flexibilidade de trabalho remoto**: Grau de flexibilidade oferecido pela empresa para trabalho remoto.
23. **Reputação que a empresa tem no mercado**: Percepção do profissional sobre a reputação da empresa no mercado.

```python
# Criando o gráfico de barra
order = df_1['Tempo de experiencia'].value_counts().index

fig, ax = plt.subplots(figsize=(15, 5))
sns.barplot(df_1['Tempo de experiencia'].value_counts(),order=order, ax=ax,  ci=0, palette = 'magma')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Tempo de Experiência')
plt.xlabel('Quantidade')
plt.ylabel('Anos')
plt.xticks(rotation=45)
plt.show()
plt.clf()
```

```python
fig, ax = plt.subplots()

ax = sns.countplot(x = 'Satisfeito', data = df_1)

bar_colors = ['darkred' if x == int('1') else 'black' for x in df_1['Satisfeito'].unique()]
for bar, color in zip(ax.containers[0], bar_colors):
    bar.set_color(color)

# Removing unnecessary spines
ax.set_title('Satisfeito(Target)')
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)

# Calculate the percentages and add them to the top of the bars
total = len(df_1['Satisfeito'])
for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        percentage = f'{round(100 * height / total, 1)}%'
        ax.text(bar.get_x() + bar.get_width() / 2, height - (0.07 * height), percentage, ha='center', va='top', color='white', fontweight='bold', fontsize=14)
plt.show()
```

1. **Distribuição Desequilibrada**:
   - A maioria das observações (64,6%) estão no grupo "1.0" (satisfeito), enquanto 25,2% estão no grupo "0.0" (não satisfeito). Isso indica que a maioria dos indivíduos ou casos no dataset estão satisfeitos.

2. **Desequilíbrio de Classe**:
   - O fato de uma classe (1.0) ter uma proporção significativamente maior do que a outra (0.0) pode ser importante ao desenvolver modelos de machine learning. Modelos treinados em datasets desequilibrados podem ser tendenciosos em favor da classe majoritária.
```python
def count_outliers(column):
    """Count the number of outliers in a column using the IQR method."""
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return ((column < lower_bound) | (column > upper_bound)).sum()

# Select numeric columns excluding 'Longitude' and 'Latitude'
numeric_cols = df_1.select_dtypes(include=['float64', 'int64'])

# Count outliers for each numeric column
outliers_count = numeric_cols.apply(count_outliers)

# Convert to DataFrame for display
outliers_count_df = outliers_count.reset_index()
outliers_count_df.columns = ['Variavel', 'Quantidade de Outliers']
outliers_count_df.sort_values(by = 'Quantidade de Outliers',ascending=False)
```

* Um alto número de outliers em certas variáveis pode indicar uma variabilidade significativa na percepção dos respondentes, sugerindo áreas que podem necessitar de atenção especial ou ajustes no gerenciamento.
